```python
def double_quote(x):
    if not(isinstance(x,(int, float, long))):
        # print '%s no' % (x,)
        try:
            x = x.replace('\'','\'\'')
        except:
            pass
    return x
# end double_quote
def wrap_in_quotes(s):
    try: 
        int(s)
        return s
    except (ValueError,TypeError):
        pass
    try:
        long(s)
        return s
    except (ValueError,TypeError):
        pass
    try:
        float(s)
        return s
    except (ValueError,TypeError):
        return '\''+ s + '\''    
    
def is_number(s):  # from the text file, it is all a string ... trying to use isinstance(x,y) will never work
    try: 
        i = int(s)
        return i 
    except (ValueError,TypeError):
        pass
    try:
        l = long(s)
        return l
    except (ValueError,TypeError):
        pass
    try:
        i = float(s)
        return i
    except (ValueError,TypeError):
        return s
```
