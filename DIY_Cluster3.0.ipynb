```python
"""
from the input file, read in and transpose to a list
first pass
"""
def parse_DIY(in_fh, out_fh):

    data = in_fh.read()
    # need this hack b/c of the way that I imported from MS Excel.
    # the line feed \r character is not recognized as a line so all the data rows are read in
    # as a single text line.

    lines = data.split('\r')
    count = 0
    proj_num = 0
    prev = ''
    col_d = {} # the column index
    sparse = [] # matrix; a list of lists
    a_li = [] # building a row of the matrix
    for line in lines:
        if truncate > 0 and count > truncate: break # debug only
        if not line: break # if the line is empty
        count += 1
        
        # parse the line
        # print line # debug only
        (line_no, proj, item) = line.split('\t')
        item=item.strip()
        
        # normalize the item text
        item = item.lower()
        item = re.sub('[\s+-]',' ',item)
        
        # add/increment item to the col_d index
        try:
            col_d[item] += 1 # increment the count
        except KeyError:
            col_d[item] = 1
        
        # generate or update the row
        # if same as the last one
        if proj == prev: 
            a_li.append(item)
        # new one
        else: 
            # update value of prev
            prev = proj
            proj_num += 1
            # add the prev to the sparse matrix
            if a_li: sparse.append(a_li)  # note: only add a_li to sparse if it is not empty; 
            # start a new one
            a_li = [proj_num, proj]
            a_li = [proj, proj]
            a_li.append(item)
        # end else
    # end for line in lines
    
    # add the last row; need an extra command b/c in the for loop, you will never trigger the IF test for the last row.
    if a_li: sparse.append(a_li)  # note: only add a_li to sparse if it is not empty; 
    
    # debug only
    # print col_index # debug only
    # print sparse # debug only
    return (col_d, sparse)

# end parse_DIY
```

```python
"""
from the input file, read in and transpose to a list
second pass

index_d:  a dictionary keyed on tools (type: strings) in the sparse_matrix.  
no error checking b/c I assume that the index_d was created by the same process and source data that created the matrix.
sparse_matrix: a sequence of sequences  sub-sequence begins with a project num, a project name, and then all other entries
are item "names"
For each row, initialize an entry in the output data struct: a sequence of sequences.
Each sequence item represents one row
Each sub-sequence is a number, a product desc, and an array of length index_d initialized to zeros.
Look-up the list and set that index value to 1.

"""
def gen_arrays(index_d, sparse_matrix):
    
    # setup variables
    index_l = index_d.keys()
    row_array = []
    count = 0
    
    for row in sparse_matrix: # this will iterate over every row in the sparse_matrix
        count += 1
        if truncate > 0 and count > truncate: break # debug 
        
        
        # initaialize a row
        a_row = []
        a_row = row[:2]
        
        # initialize the vector to zeros
        z_array = numpy.zeros(len(index_l))
        
        # update the vector index for every item in the project 
        for item in row[2:]:
            z_array[index_l.index(item)] = 1
        
        # add the z_array) to the row
        a_row.append(z_array)
        
        # add the row to the output
        row_array.append(a_row)
    # end for row in sparse matrix
        
    # print index_l # debug
    # print row_array # debug
    return (index_l, row_array)

        
    

# end gen_arrays
```

```python
def print_array(index_l, row_array, ofh):
    # initialize header
    line = '%s\t%s\t' % ('proj_num','proj_name')
    for label in index_l:
        line = line + '\t' + label
    line = line + '\n'
    # end for label

    # print line # debug
    ofh.write(line)
    
    # data rows
    for row in row_array:
        # initialize a line
        line = '%s\t%s\t' % (str(row[0]),row[1]) # assumes the row[0] item is an integer
        
        for val in row[2]: # assumes the row[2] item is an array; no error checking; could add TRY ERROR on type
            line = line + '\t' + str(val)
        line = line + '\n'
        
        # print line # debug
        ofh.write(line)
    # end for row in row_array
    

# end print_array(index_l, row_array, ofh)
    


```
