```python
"""
update.
return a list of dictionaries that are processed for output.
"""

def parse_DIY_bread(in_fh, out_fh):
    output_d = {} # output
    count = 0
    digits_d = {} # anything w/ numbers
    punct_d = {}  # anything w/ something other than num and char
    one_d = {}    # one token (no whitespace)
    two_d = {}    # two tokens (one whitespace)
    more_d = {}   # more
    while 1:
        line = in_fh.readline()
        if not line: break # if the line is empty
        if count > test and test > 0: break
        count += 1
        # line = line.strip()
        line = line.lower() # lower case
        line = line.replace('&amp;','&')
        parse = line.split('\t')
        project = parse[1]
        item = parse[2]
        item.strip() # clear leading/trailing whitespace
        # prob just apply Porter Stemming here.
        if item[-2:] == 'es': item = item[:-2] # cut the last two characters off if it is an 's'   
        if item[-1:] == 's': item = item[:-1] # cut the last character off if it is an 's'
        breadcrumb = parse[3].split(',')
        breadcrumb = map(lambda x: x.strip(),breadcrumb)
        #         print parse[:3] # slice off the end w/ breadcrumb
        #         print breadcrumb
        #         print breadcrumb[len(breadcrumb)-1]
        
        
        if contains_digits(item):
            try:
                digits_d[item].append(project)
            except KeyError:
                digits_d[item] = [project]
        elif contains_non_alphanum(item):
            try:
                punct_d[item].append(project)
            except KeyError:
                punct_d[item] = [project]
        else:
            # counting num tokens by looking at num whitespace
            x = num_whitespace(item) # returns a list of matches or else None 
            if not x: 
                try:
                    one_d[item].append(project)
                except KeyError:
                    one_d[item] = [project]
            elif len(x) == 1:
                try:
                    two_d[item].append(project)
                except KeyError:
                    two_d[item] = [project]
            else:
                try:
                    more_d[item].append(project)
                except KeyError:
                    more_d[item] = [project]
        # end else counting num tokens

        
        
    # end 1 reading in the file
        
#     print 'DIGIT NAMES\n'
#     for s in digits_d.keys(): print '%s\n' % (s,)
#     print 'PUNCT NAMES\n'
#     for s in punct_d.keys(): print '%s\n' % (s,)
#     print 'ONE TOKEN NAMES\n'
#     for s in one_d.keys(): print '%s\n' % (s,)
#     print 'TWO TOKEN NAMES\n'
#     for s in two_d.keys(): print '%s\n' % (s,)
#     print 'MORE TOKEN NAMES\n'
#     for s in more_d.keys(): print '%s\n' % (s,)
#     output_l = [digits_d, punct_d, one_d, two_d, more_d]
    output_d['DIGIT_NAMES'] = digits_d  # need to look at breadcrumbs, not a good way to parse
    output_d['PUNCT_NAMES'] = punct_d   # only 52 of these; parse somehow?
    output_d['ONE_TOKEN_NAMES'] = one_d
    output_d['TWO_TOKEN_NAMES'] = two_d
    output_d['MORE_TOKEN_NAMES'] = more_d # name branded goods.
    

    return output_d

# end parse_DIY_bread
        
```
